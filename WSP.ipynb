```python
import numpy as np
import os
import cv2
dirs=os.listdir("C:\\Users\\Shrey Malvi\\Desktop\\dataset-resized")
print(dirs)
objects=[]
labels=[]
x={'0':2,'1':0,'2':4,'3':1,'4':3,'5':5}
i=0
for file in dirs:
    print(file)

    c=0
    if(i<6):
        for images in os.listdir("C:\\Users\\Shrey Malvi\\Desktop\\dataset-resized\\"+file):
            img1=cv2.imread("C:\\Users\\Shrey Malvi\\Desktop\\dataset-resized\\"+file+"\\"+images)
            print(img1.shape)
            img1=cv2.resize(img1,(128,128))
            print(img1.shape)
            if(c%50==0):
                print(c)
            c=c+1
            objects.append(np.asarray(img1))
            labels.append(x[str(i)])
    i=i+1
obj=np.stack( objects )
lab=np.stack( labels )
#np.save('objects_data',obj)
#np.save('labels_data',lab)
```

```python
print(obj.shape)
```

```python
print(lab.shape)
```

```python
labels=to_categorical(lab)
```

```python
print(labels.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(obj,labels, test_size=0.2)
```

```python
from keras.applications import VGG16
#Load the VGG model
vgg_conv = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
```

```python
for layer in vgg_conv.layers[:]:
    layer.trainable = False

for layer in vgg_conv.layers:
    print(layer, layer.trainable)
```

```python
model = models.Sequential()
 
# Add the vgg convolutional base model
model.add(vgg_conv)
 
# Add new layers
model.add(layers.Flatten())
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dropout(0.4))
model.add(layers.Dense(6, activation='softmax'))
model.summary()
```

```python

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
model.fit(X_train, y_train, epochs=30, batch_size=50)
```
